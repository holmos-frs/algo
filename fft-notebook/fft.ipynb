# FFT Beispiel

Eine Fourier Transformation von einem erzeugten Streifenmuster

```python
%matplotlib notebook

import ipywidgets as widgets

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy.misc import imread

selecter = widgets.Select(
    options=['bild.jpg', 'streifen_dicker.png', 'streifen_gebogen.png', 'streifen_klein.png', 'streifen_klein_riesig1.png', 'streifen_schraeg.png'],
    value='bild.jpg',
    # rows=10,
    description='Streifenbild:',
    disabled=False
)

aktuellesbild = 'bild.jpg'

def show_picture(value):
    aktuellesbild = value['new']
    print(aktuellesbild)
    
    # Load image from file
    image = (imread(aktuellesbild, mode='L'))
    print(image.shape)
    width = image.shape[0]
    height = image.shape[1]
    
    # fft
    fft = np.fft.fft2(image)
    fshift = np.fft.fftshift(fft)
    
    plt.clf()
    plt.subplot(121)
    plt.imshow(image, cmap='gray')
    
    plt.subplot(122)
    plt.imshow(np.abs(fshift+1), cmap='gray')
    #plt.show()
    
    # Fouriere Transformation
    

display(selecter)
selecter.observe(show_picture, names='value')

```
